```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
%matplotlib inline 
```

```python
def mandelbrot(h, w, maxit=20, r=2):
    """Returns an image of the Mandelbrot fractal of size (h,w)."""
    x = np.linspace(-2.5, 1.5, 4*h+1)
    y = np.linspace(-1.5, 1.5, 3*w+1)
    A, B = np.meshgrid(x, y)

    C = A + B*1j
    z = np.zeros_like(C)
    divtime = maxit + np.zeros(z.shape, dtype=int)

    for i in range(maxit):
        z = z**2 + C
        diverge = abs(z) > r                    # who is diverging
        div_now = diverge & (divtime == maxit)  # who is diverging now
        divtime[div_now] = i                    # note when
        z[diverge] = r                          # avoid diverging too much

    return divtime
```

```python
def mandelbrot2(h, w, maxit=20, r=2, window=(-2.5,1.5,-1.5,1.5)):
    """Returns an image of the Mandelbrot fractal of size (h,w)."""
    x = np.linspace(window[0], window[1], 4*h+1)
    y = np.linspace(window[2], window[3], 3*w+1)
    A, B = np.meshgrid(x, y)

    C = A + B*1j
    z = np.zeros_like(C)
    divtime = maxit + np.zeros(z.shape, dtype=int)

    for i in range(maxit):
        z = z**2 + C
        diverge = abs(z) > r                    # who is diverging
        div_now = diverge & (divtime == maxit)  # who is diverging now
        divtime[div_now] = i                    # note when
        z[diverge] = r                          # avoid diverging too much

    return divtime
```

```python
plt.clf()
window = (-2.0,0.75,-1.5,1.5)
plt.imshow(mandelbrot2(400, 400, window=window));
```

```python
def show(data):
    img = Image.frombytes('1', data.shape[::-1], np.packbits(data, 1))
    img.show()
```

```python
n = 100
maxiter = 100

M = np.zeros([n, n], np.uint8)
xvalues = np.linspace(-2, 2, n)
yvalues = np.linspace(-2, 2, n)

for u, x in enumerate(xvalues):
    for v, y in enumerate(yvalues):
        z = 0
        c = complex(x, y)
        for i in range(maxiter):
            z = z*z + c
            if abs(z) > 2.0:
                M[v, u] = 1
                break

plt.plot(M)
plt.title('A simple chirp')
plt.show()
```
